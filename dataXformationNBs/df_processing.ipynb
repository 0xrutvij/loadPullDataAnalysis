```python
csvName = "UTD_LP_File_1.csv"
pickleName = "utd_loadpull.pkl"
#df_csv = pd.read_csv(csvName) #filepath + 
df = pd.read_pickle(pickleName) #filepath +
```

```python
###########this is the excel read into a DF
#df.head(5)


################################################
########      DATA EXPLORATION      ############
################################################

df1 = df[(df['harmonic'] == 1)]
display(df1.describe())

meanPout = 0.031173

df1 = df1[df1['Pout'] >= meanPout]

#print(df1.head())

import plotly.express as px

#lColors = ['Blue','Red','Yellow','Black']

fig = px.scatter(df1, x="Pout", y='Pin', title="Pout vs Pin", color='power')

fig.show()

```

```python
#Filter By Power == 6
power_harm = df[(df['power'] >= 6) & (df['harmonic'] == 1)]
power_harm = power_harm[['gammaTuple','Gain', 'r', 'jx', 'PAE', 'power', 'Pin', 'Pout']]
g1 = []
for i, row in power_harm.iterrows():
    gT = row['gammaTuple']
    g1.append(gT[0])

power_harmC = power_harm.copy()
power_harmC['gamma1'] = g1

import plotly.graph_objects as go

fig = go.Figure(data =
    go.Contour(
        z=power_harm['Pout'],
        x=power_harm['r'], # horizontal axis
        y=power_harm['jx'] # vertical axis
    ))
fig.show()
#print(power_harmC)
#filter on harmonics, 2 and 3 don't have all column values display only harmonic 1
```

```python
# scatter plot in-built df plot
#power_harmC.plot(x='gamma1', y='PAE', style='o')
#lColors = ['Blue','Red','Yellow','Black']

fig = px.scatter(power_harmC, x="gamma1", y='Pout', title="Pout vs GammaL1", color='power')

fig.show()
```

```python
import plotly.express as px

#lColors = ['Blue','Red','Yellow','Black']

fig = px.scatter(power_harmC, x="gamma1", y='PAE', title="PAE vs GammaL1", color='power')

fig.show()
```

```python
fig = px.scatter(power_harmC, x="gamma1", y='Pin', title="Scatter Plot", color='power')

fig.show()


gStd = power_harmC['Gain'].std()
mean = power_harmC['Gain'].mean()
upL, lowL = mean+gStd, mean-gStd

power_harmC1 = power_harmC[(power_harmC['Gain'] > (lowL/2)) & (power_harmC['Gain'] < (upL*3/5))]

fig = px.scatter(power_harmC1, x="gamma1", y='Gain', title="Pin vs GammaL1", color='power')

fig.show()
```

```python
fig = px.scatter_3d(power_harmC1, x="PAE", y='gamma1', z='Gain', title="Gain vs PAE", color='power', size_max=10)

fig.show()
```

```python
import math

##Closest I could get to a Smith Chart


dfC = power_harmC1.copy()

thet = lambda r,x: math.degrees(math.atan(r/x))
dfC['theta'] = dfC.apply(lambda x: thet(x.r, x.jx), axis=1)

rootSqSum = lambda r,x: math.sqrt(r**2+x**2)
dfC['Z'] = dfC.apply(lambda x: rootSqSum(x.r, x.jx), axis=1)

fig = px.scatter_polar(dfC, r="Z", theta="theta")
fig.show()
```

```python
import plotly.graph_objects as go
fig = go.Figure()
fig.add_trace(go.Scatterpolar(
        r = dfC['PAE'],
        theta = dfC['theta'],
        mode = 'markers',
    ))
fig.show()
```
